# K-Nearest Neighbour(KNN) -

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
```

```python
dataset=pd.read_csv("train.csv")
```

```python
dataset.head(2)
```

```python
dataset.columns 
```

```python
dataset.isna().sum()
```

```python
dataset1=dataset.drop(["Name","Ticket","Cabin"],axis=1)
```

```python
dataset1.columns
```

```python
le=LabelEncoder()
le.fit(dataset1["Sex"])
```

```python
dataset1["Sex"]=le.transform(dataset1["Sex"])
```

```python
print(le.classes_)
```

```python
le.fit(dataset1["Embarked"])
```

```python
print(le.classes_)
```

```python
dataset1["Embarked"]=le.transform(dataset1["Embarked"])
```

```python
print(le.classes_)
```

```python
from sklearn import neighbors
```

```python
KNN_dataset=dataset1
```

```python
KNN_dataset.info()
```

```python
KNN_dataset.drop_duplicates()
```

```python
KNN_dataset.info()
```

```python
y=KNN_dataset["Pclass"]
X=KNN_dataset.drop(["Pclass","PassengerId"],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
knn=neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train).score(X_test,y_test)*100
```

```python
y_pred=knn.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
def K_near():
    for i in range(1,268):
        knn=neighbors.KNeighborsClassifier(n_neighbors=i)
        acc_score=knn.fit(X_train,y_train).score(X_test,y_test)*100
        Acc=np.array(acc_score)
        print("The accuracy of the model we developed with respect to k=", i, "value is:\n",Acc,"%")
        y_pred=knn.predict(X_test)
        print("The respective confusion matrix corresponding to K-value is:\n", confusion_matrix(y_test,y_pred))
        i=i+1
K_near()
```

# Conclusion - 

**Considering all the K-values possible from 1 to 267 we have found that the model has the maximum value when K=1 having the Accuracy Score of 89.13857677902621 %**

# Support Vector Machine(SVM) - 

```python
KNN_dataset.info()
```

```python
SVM_dataset=KNN_dataset
```

```python
SVM_dataset.head(2)
```

```python
SVM_dataset.columns
```

```python
Attributes = ['Pclass', 'Survived', 'Sex', 'SibSp', 'Parch', 'Embarked']
```

```python
from sklearn import svm
def SVM(x,y):
    x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
    clf = svm.SVC(gamma = 0.01)
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print("Accuracy_score:",accuracy_score(y_test, y_pred))
    print("Confusion-matrix:")
    print(confusion_matrix(y_test, y_pred))
```

```python
for DV in Attributes:
    y = SVM_dataset[DV]
    x = SVM_dataset.drop(DV,axis=1)
    print("")
    print(DV,"Vs Remaining Dv")
    SVM(x,y)
```
